# SVM 

### It is supervised machine learning algorithm for Classification and Regression to handling linear or nonlinear
### Objective is to find the optimal hyperplane 
### Two marginal plane will be created and parallel to the hyperplane
### The distance from marginal 1 to hyperplane and The distance from marginal 2 to hyperplane should be equal.
### The distance from marginal 1 to marginal 2 should be maximum which is know as hard marginal plane
