### Decision Tree :
### It is supervised machine learning algorithm for classification and regression
### It has a hierarchical tree structure consisting of a root node, Decision nodes, and leaf or Terminal nodes
### Features Split will happens based on the information gain
### Entropy and gini index technique is using to check the purity of leaf node

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
iris.data
```

```python
iris.target
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier(criterion='entropy')
classifier=DecisionTreeClassifier(criterion='entropy',max_depth=2)
classifier=DecisionTreeClassifier(criterion='gini')
classifier=DecisionTreeClassifier(criterion='gini',max_depth=2)
classifier.fit(iris.data,iris.target)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(classifier,filled=True)
```
