### 4.1) InterQuartile Range

#### IQR = Q3 - Q1
#### Q1 is 25 percentile and Q3 is 75 percentile
#### 1) How to find the 25 percentile ?
#### Note : 25% percentile is similar to percentage formuale

#### value = (Percentile / 100) * (n+1)

```python
rangelist=[1,2,3,4,5,6,7,3,4,5,6,11]
rangelist.sort()
percentile25_index=(25/100)*(len(rangelist)+1)
percentile25_index=round(percentile25_index)
print("The Value of 25 percentile is {}".format(rangelist[percentile25_index]))
```

#### 2) How to find the 4 is what percentile ?
#### value = (index  / n) * 100

```python
print(rangelist)
value=(rangelist.index(4)/len(rangelist))*100
print("The Value of 4 is {} percentile".format(value))
print("The value of 4 is greater than 33.4 percentage")
```
